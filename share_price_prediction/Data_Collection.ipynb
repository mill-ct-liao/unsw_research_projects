```python
import numpy as np
import pandas as pd
import yfinance as yf
from utils import *

config = read_config('config.yml')
```

```python
asx200_list = pd.read_excel(config['asx200_list'])

start_date = config['start_date']
end_date = config['end_date']

codes = list(asx200_list.Code)

for code in codes:
    try:
        company_prices = yf.Ticker(code + '.AX').history(start=start_date, end=end_date)
        company_prices['code'] = code
        historical_prices = pd.concat([historical_prices, company_prices])
    except:
        print('begin')
        historical_prices = yf.Ticker(code + '.AX').history(start=start_date, end=end_date)
        historical_prices['code'] = code
```

```python
data_value_counts = historical_prices.code.value_counts()
companies_5y = data_value_counts[data_value_counts==config['5y_companies']].index.tolist()
companies_5y_historical_prices = historical_prices[historical_prices.code.isin(companies_5y)]
companies_5y_historical_prices = companies_5y_historical_prices.loc[:,['Open','High','Low','Close','Volume','code']]
```

```python
companies_5y_historical_prices.to_excel(config['asx200_historical_price'])
```
